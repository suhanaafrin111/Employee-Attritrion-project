```python
# Exploratory Data Analysis (EDA)
```

```python
# Attrition Count
sns.countplot(x='Attrition', data=df)

plt.title("Employee Attrition")

plt.show()
```

```python
# Department Distribution
sns.countplot(
    x='Department',
    data=df
)

plt.xticks(rotation=30)

plt.show()
```

```python
# Gender Distribution
sns.countplot(
    x='Gender',
    data=df
)

plt.show()
```

```python
# Age Distribution
plt.hist(df['Age'], bins=20)

plt.xlabel("Age")

plt.ylabel("Employees")

plt.show()
```

```python
# Monthly Income Distribution
sns.histplot(df['MonthlyIncome'], kde=True)

plt.show()
```

```python
# correlation heatmap
plt.figure(figsize=(15,10))

sns.heatmap(
    df.corr(numeric_only=True),
    cmap='coolwarm'
)

plt.show()
```

```python
# HR KPIs
```

```python
# Total Employees
total_employees = df.shape[0]

print(total_employees)
```

```python
# Employees Left
employees_left = df[df['Attrition']=='Yes'].shape[0]

print(employees_left)
```

```python
# Active Employees
active = df[df['Attrition']=='No'].shape[0]

print(active)
```

```python
# Attrition Rate
attrition_rate = employees_left / total_employees * 100

print(round(attrition_rate,2))
```

```python
# Department-wise Attrition
dept = pd.crosstab(
    df['Department'],
    df['Attrition']
)

dept
```

```python
dept.plot(kind='bar')

plt.title("Department vs Attrition")

plt.show()
```
